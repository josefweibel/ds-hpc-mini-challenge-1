```python
import timeit as timeit
import numpy as np
import cProfile
from pstats import SortKey
from rating_producer import publish as publish_ratings
from rating_producer import read_file as read_ratings_file
from rating_producer import producer as rating_message_producer
from movie_data_producer import publish as publish_movies
from movie_data_producer import read_file as read_movies_file
from movie_data_producer import producer as movie_data_message_producer
from mean_rating_producer import producer as mean_rating_producer
from movie_data_sink import processor as movie_data_processor
```

# Code Performance Analyse

## Data Generators

```python
n = 100
def setup():
	global ratings_file
	ratings_file = read_ratings_file()

def run():
	global ratings_file
	publish_ratings(ratings_file, n)

results = timeit.repeat(run, setup, repeat = 3, number = 10)
print(f'Publishing {n} ratings: mean = {np.min(results)}s +/-{np.std(results)}s')
```

```python
n = 100
def setup():
	global movies_file
	movies_file = read_movies_file()

def run():
	global movies_file
	publish_movies(movies_file, n, wait_time = 0)

results = timeit.repeat(run, setup, repeat = 3, number = 10)
print(f'Publishing {n} movies: mean = {np.min(results)}s +/-{np.std(results)}s')
```

## Data Processors

```python
with cProfile.Profile() as mean_rating_producer_report:
	mean_rating_producer.run(1000)
```

```python
mean_rating_producer_report.print_stats(sort = SortKey.TIME)
```

```python
mean_rating_producer_report.dump_stats('mean_rating_producer-improved.profile')
```

```python
with cProfile.Profile() as movie_data_processor_report:
	movie_data_processor.run(1000)
```

```python
movie_data_processor_report.print_stats(sort = SortKey.CUMULATIVE)
```

```python
movie_data_processor_report.dump_stats('movie_data_processor-improved.profile')
```
